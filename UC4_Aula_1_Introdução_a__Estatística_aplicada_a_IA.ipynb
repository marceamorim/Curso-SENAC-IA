# O que é Estatística?

Estatística é a área que trabalha com:

- Coleta de dados
- Organização de dados
- Análise de dados
- Tomada de decisão

---

## Exemplos

- Média de notas → aprovação
- Consumo → recomendação
- Padrões → detecção de fraude

O objetivo da estatística é transformar dados em informação para apoiar decisões.

---

## Dado x Informação

### Dado
É um valor bruto, sem contexto.

Exemplos:
- 25
- 3000
- "DF"

### Informação
É o dado interpretado com significado.

Exemplos:
- 25 anos de idade
- salário de 3000 reais
- mora em São Paulo

---

## População e Amostra

### População
É o conjunto total de elementos que queremos estudar.

Exemplo:
- todos os alunos de uma escola
- todos os clientes de uma empresa

---

### Amostra
É uma parte da população utilizada para análise. Um subconjunto da populacao

Exemplo:
- 30 alunos de uma turma
- 100 clientes selecionados

---

### Tipos de amostragem (o mais importante)
1. Amostragem aleatória simples
* Todos têm a mesma chance de ser escolhidos

Exemplo:

sortear nomes de uma lista

(É o método mais correto)

----
2. Amostragem por conveniência

(Escolhe quem está mais fácil)

Exemplo:

entrevistar só quem está na sala

Problema: pode gerar viés

3. Amostragem estratificada

(Divide a população em grupos e pega um pouco de cada)

Exemplo:

homens e mulheres
diferentes idades

(Muito usada em pesquisa)

-----
4. Amostragem sistemática

(Seleciona de forma regular)

Exemplo:

a cada 10 pessoas, escolhe 1

| Etapa | Uso da estatística |
|------|------------------|
| Coleta | Organização de dados |

Sem estatística, a IA não consegue aprender.

# Tipos de Dados

## Variável

Variável é qualquer característica que pode ser observada ou medida.

Exemplos:
- idade
- salário
- cidade
- nota

---



## Variáveis Qualitativas

- Não são numéricas (ou não fazem sentido calcular)
- Representam categorias

Ex:
- Nome
- Cor
- Cidade
- Estado civil

---

## Variáveis Quantitativas

- São numéricas
- Permitem cálculos

Ex:
- Idade
- Salário
- Altura
- Quantidade

## Variáveis Quantitativas (Subtipos)

###  Discretos
- Valores inteiros
- Contáveis

Ex:
- Número de filhos
- Número de compras

---

###  Contínuos
- Podem ter qualquer valor
- Incluem decimais

Ex:
- Altura
- Peso
- Temperatura

## Variáveis Qualitativas (Subtipos)

### Nominais
- Não possuem ordem
- - Ex: cor, cidade, gênero

### Ordinais
- Possuem ordem
- - Ex: escolaridade, satisfação, ranking
